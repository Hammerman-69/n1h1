```python
# Analyse the dataset columns for Null Values by creating a heatmap where null=true is the heatmap value
plt.subplots(figsize=(20,15))
sns.set(font_scale=2)
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Checking Columns for Number of Null Records and Dropping where it is >=50%

colRemovalList = []

for colName in train:
    nullRecordCount = len(train[(train[colName].isna()==True)])#Count the number of Null values in the column
    checkVal = nullRecordCount/ (train[colName].count() + nullRecordCount)#Get the Full Count and get the ratio of nulls to populated                    
    
    if (checkVal >=.5): # more than 50% nulls, drop the column
        colRemovalList.append(colName)
        print(f'{colName} to be deleted, with a null record count of {nullRecordCount} out of {train[colName].count() + nullRecordCount} total rows\n')     

#Drop the column contents in the list
train.drop(colRemovalList, axis =1, inplace=True)

#Confirm the columns are now removed from the dataframe
#train.info()
```

```python
#For Each of Column with Nulls calculate the distribution freq to populate null values.

for colName in train:
    dist = train[colName].value_counts(normalize=True)#Distribution of populated values in column
    #print(dist)
    
    #Get the Rows when value is not set
    nan_rows  = train[colName].isna() == True
    #Using the distribution Freq populate the Nulls
    train.loc[nan_rows,colName] = np.random.choice(dist.index, size=len(train[nan_rows]),p=dist.values)

    #DEBUGGING STUFF------------------------
    #Check its Null
    #print(len(train[(train[colName].isna()==True)]))
    
    #Check the Dist
    #print(train[colName].value_counts(normalize=True))
    #print(train[colName].value_counts())
    #print("Next Loop") 

print(f"There are now {train.isnull().values.sum()} Null Records left in the Data Set")
```

```python
#categoricals column need to be converted to a format the Model can understand
df_categories = train.select_dtypes(include=['object']).copy()
df_categories.info()
```

```python
#Remove the Age Buckets
#Convert to a Set to remove the duplicates
ageBucketsSet = set(df_categories['age_group'].values.tolist())
ageBucketMap = {}

#Need to detail with the text format of this column
for ageRange in ageBucketsSet:  
    z = ageRange.replace("Years", "").replace('+', '').strip()
    
    if (z.find('-') != -1):
        startRange, endRange = z.split('-')
        startRange = startRange.strip()
        endRange = endRange.strip()
        AgeValue = round(((int(startRange)+int(endRange))/2))
    else:
        AgeValue = int(z)
    
    #Add the Key Value Pairs to the dict for mapping the new values in the DataFrame    
    ageBucketMap[ageRange] = AgeValue
    
averageAge = df_categories["age_group"].map(ageBucketMap)
averageAge.name = 'averageAge'

print(round(np.mean(averageAge)))#inspect the averageAge across the dataset
```

```python
#Function to remove Catogoricals and replace with Binary
def remapcategoricals(dataframe, appendLabel, simpleDrop):
    temp_df = pd.get_dummies(dataframe, drop_first = simpleDrop) #simpleDrop True/False value passed in
    
    #rename the columns so its still obvious what they represent  
    for col in temp_df.columns:
        temp_df.rename(columns={col: f'{appendLabel} {col}'}, inplace=True)
    
    return temp_df
```

```python
#Quick view of the category dataframe before its changed in remapcategoricals
df_categories.head(10)
```

```python
#Convert Categoricials To Binary
originalColList = ['age_group'] # age_group done above with the Average Age Buckets

education_df = remapcategoricals(df_categories['education'], 'Education', False)
originalColList.append('education')

race_df = remapcategoricals(df_categories['race'], 'Race', False) 
originalColList.append('race')

sex_df = remapcategoricals(df_categories['sex'], 'Sex', True)
originalColList.append('sex')

income_poverty_df = remapcategoricals(df_categories['income_poverty'], 'Income_Poverty', False)
originalColList.append('income_poverty')

marital_status_df = remapcategoricals(df_categories['marital_status'], 'Marital_Status', False)
originalColList.append('marital_status')

rent_or_own_df = remapcategoricals(df_categories['rent_or_own'], 'Rent_Or_Own', True)
originalColList.append('rent_or_own')

employment_status_df = remapcategoricals(df_categories['employment_status'], 'Employment_Status', True) 
originalColList.append('employment_status')

hhs_geo_region_df = remapcategoricals(df_categories['hhs_geo_region'], 'Hhs_Geo_Region', True)
originalColList.append('hhs_geo_region')

census_msa_df = remapcategoricals(df_categories['census_msa'], 'Census_Msa', True) 
originalColList.append('census_msa')

employment_industry_df = remapcategoricals(df_categories['employment_industry'], 'Employment_Industry', True) 
originalColList.append('employment_industry')
```

```python
#drop a column so they are not perfect predictors of each other, other columns did this in the function, it logical sense to pick the column for these categories, wanted lowest val in seq 
education_df.drop('Education < 12 Years', axis =1, inplace=True)
race_df.drop('Race Black', axis =1, inplace=True)
income_poverty_df.drop('Income_Poverty Below Poverty', axis =1, inplace=True)
marital_status_df.drop('Marital_Status Not Married', axis =1, inplace=True)
```

```python
#Add the new binary columns and drop the original columns and the identifier column
dfList = [train,averageAge, education_df ,race_df,sex_df,income_poverty_df,marital_status_df ,rent_or_own_df ,employment_status_df ,hhs_geo_region_df,census_msa_df,employment_industry_df]
linearRegressionModel_df = pd.concat(dfList, axis =1)

originalColList.append('respondent_id')
for col in originalColList:
    linearRegressionModel_df.drop(col, axis=1, inplace=True)
      
# All Categorical Data should now be removed from the dataframe, have a quick peek
linearRegressionModel_df.head(10)
```

```python
#Split the dataset by X and y, with y to hold h1n1_vaccine values
from sklearn.model_selection import train_test_split

y = linearRegressionModel_df['h1n1_vaccine']#isolcate the Target Label
X = linearRegressionModel_df.drop('h1n1_vaccine', axis = 1)#Drop the Target Label, this is the prediction column

#Split the data into the Train Test Split with a test size of 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Data Set used for split
linearRegressionModel_df.head(10)
```

```python
#Test Split, double checking the h1n1_vaccine is gone from this data set
X_test.head(10)
```

```python
#Run the Linear Regression Model
from sklearn.linear_model import LogisticRegression
logisticModel = LogisticRegression()
logisticModel = LogisticRegression(solver='liblinear', max_iter=1000)
logisticModel.fit(X_train, y_train)
logisticModelPredictions = logisticModel.predict(X_test)
```

```python
#Review the Results, show the Accuracy and the Confusion Matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

accuracy = '{:.3f}'.format(float(accuracy_score(y_test, logisticModelPredictions)))
plt.figure(figsize=(9,9))
sns.heatmap(confusion_matrix(y_test, logisticModelPredictions), annot=True, fmt=".3f", linewidth=.5, square=True, cmap="Blues_r")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title(f'Logistic Regression, Accuracy = {accuracy}')
```

```python
#View the classification report for the Model results
from sklearn.metrics import classification_report
print(f'Logistic Regression - Classification Report:\n {classification_report(y_test, logisticModelPredictions)}')
```
